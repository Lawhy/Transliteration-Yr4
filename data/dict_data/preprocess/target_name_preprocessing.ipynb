常规汉字unicode范围是4E00-9FA5， 共20902字符。

```python
ch_pa = re.compile(r'^[\u4E00-\u9FA5]+$')
re.findall(ch_pa, '霍华德')
```

```python
count = 0
nch_inds = []
for index, dp in raw_en.iterrows():
    tgt = dp['target_name']
    if not re.findall(ch_pa, tgt):
        print(tgt)
        nch_inds.append(index)
        count += 1
```

```python
print(count)
```

```python
all(raw_en.loc[nch_inds] == raw_en.iloc[nch_inds])
```

```python
nch =  raw_en.loc[nch_inds]
nch.head(10)
```

```python
nch.describe()
```

```python
bra_pa = re.compile('^[\u4E00-\u9FA5]+?\(.+?\)$') 
bra_inds = []
colon_inds = []
el_inds = []
for index, dp in nch.iterrows():
    tgt = dp['target_name']
    if not re.findall(bra_pa, tgt):
        if ';' in tgt:
            colon_inds.append(index)
        else:
            print(tgt)
            el_inds.append(index)
    else:
        bra_inds.append(index)
```

```python
print(len(el_inds))
print(len(el_inds) + len(bra_inds) + len(colon_inds))
```

先处理非$\textbf{分号}$或者$\textbf{纯括号}$的情况，即el_inds

```python
el = raw_en.loc[el_inds]
el
```

```python
dots_pa = re.compile('^([\u4E00-\u9FA5]+?)\.$')
# remove dots and brackets
for index, dp in el.iterrows():
    tgt = dp['target_name']
    if re.findall(dots_pa, tgt):
        dp['target_name'] = re.findall(dots_pa, tgt)[0]
```

```python
bra_sp_pa = re.compile(r'^([\u4E00-\u9FA5]+?) \(.+?\)$')
# remove bracket with spaces
for index, dp in el.iterrows():
    tgt = dp['target_name']
    if re.findall(bra_sp_pa, tgt):
        dp['target_name'] = re.findall(bra_sp_pa, tgt)[0]
```

```python
inverted_comma_inds = []
en_pa = re.compile('^[a-zA-Z]+$')
en_pa_2 = re.compile('^[a-zA-Z\']+$')
for index, dp in raw_en.iterrows():
    source = dp['source_name']
    if not re.findall(en_pa, source):
        if not re.findall(en_pa_2, source):
            print(' --- ')
        else:
            inverted_comma_inds.append(index)
```

```python
raw_en.loc[inverted_comma_inds].to_excel('names_with_inverted_comma.xlsx', encoding='UTF-8-sig')
```

```python
for index, dp in raw_en.loc[inverted_comma_inds].iterrows():
    src = dp['source_name']
    src = src.replace('\'', '')
    raw_en.loc[index]['source_name'] = src
```
