```python
import re
import pandas as pd
import numpy as np
```

```python
raw = raw.rename({'第一部分 外国姓名（日、朝、越等使用和曾使用汉字国家人名除外）': 'source_name', 'Unnamed: 1': 'origin', 'Unnamed: 2':'target_name'}, axis=1)
```

字典里共有672039个词目

```python
raw = raw.astype(str)
```

```python
raw_en = raw[raw['origin'].str.contains('英')]
```

```python
raw_en.describe()
```

```python
for d in raw_en['source_name']:
    d = d.strip()
    if re.findall(re.compile('[^a-zA-Z]'), d):
        print(d)
```

```python
count = 0
for index, dp in raw_en.iterrows():
    src = dp['source_name']
    if re.findall(re.compile('[^a-zA-Z]'), src):
        print(src)
        src = src.replace(' ', '')
        src = src.replace('<sup>′</sup>', '\'')
        dp['source_name'] = src
        count += 1
```

```python
print(count)
```

352 srouce names containing non-English letters, keep those with spaces and replace $<sup>′</sup>$ with $'$.

```python
count = 0
for index, dp in raw_en.iterrows():
    src = dp['source_name']
    if re.findall(re.compile('[^a-zA-Z]'), src):
        if not re.match(r'^[a-zA-Z\']+$', src):
            print(src)
            raw_en = raw_en.drop(index=index)
            count += 1
```

```python
print(count)
```

51 exceptional names needed to be removed.

```python
raw_en.describe()
```

####  Make sure the first step of data extraction (source name) is done right.

```python
count = 0
for index, dp in raw_en.iterrows():
    src = dp['source_name']
    if re.match(r'^[a-zA-Z\']+$', src):
        if '\'' in src:
            print(src)
            count += 1
    else:
        print('Error!', src)
```

```python
print(count)
```

####  May be useful afterwards (directly processing the old version of people.txt)

```python
with open('People.txt', 'r', encoding='UTF-8-sig') as f:
    data = f.readlines()
```

```python
raw_pattern = r'(.+?)\t(.+?)\t(.+)\n'
raw_df = pd.DataFrame(columns=['English', 'Chinese'])
en_names = []
ch_names = []
sources = []

for line in data:
    pairs = re.findall(raw_pattern, line)
    if pairs:
        en_names.append(pairs[0][0])
        ch_names.append(pairs[0][2])
        sources.append(pairs[0][1])
    else:
        print(pairs)
```

```python
raw_df['English'] = en_names
raw_df['Chinese'] = ch_names
raw_df['Source'] = sources
raw_df.describe()
```

```python
raw_df[raw_df['English'] == 'Finola']
```

### Remove noise '(教名...)' 

```python
bracket_pattern = r'(.+)?\((.+)\)'
counts = dict()
temp = 0
temp1 = 0
temp2 = 0
for index, dp in raw_df.iterrows():
    ch = dp['Chinese']
    if re.findall(bracket_pattern, ch):
        temp += 1
        pairs = re.findall(bracket_pattern, ch)
        content_in_bracket = pairs[0][1]
        if '教名' in content_in_bracket:
            temp1 += 1
            dp['Chinese'] = pairs[0][0]
            # print(pairs[0][0])
            if not pairs[0][0] in raw_df['Chinese']:
                temp2 += 1
                print(dp['English'], dp['Chinese'], dp['Source'])
counts['first'] = [temp, temp1, temp2] 
```

```python
print(counts)
raw_df.describe()
```

去掉“教名”以后，unique Chinese names从43474减少到43144，共少了330个附加词“教名”。
